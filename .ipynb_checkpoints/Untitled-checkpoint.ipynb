```python
import csv

# 创建词典
dictionary={}
# 指定词典
with open('E:/大三下/中南大学自然语言处理实验指导书（一）/nlpproj/chinese_word_freq_list.txt', 'r', encoding='utf-8') as fd:
     lines=fd.readlines()
     for line in lines:
          l=line.strip().split()
          dictionary[f'{l[1]}']=l[2]

# 根据train.csv扩充词典
with open('E:/大三下/中南大学自然语言处理实验指导书（一）/nlpproj/train.csv', 'r', encoding='utf-8') as ft:
     reader = csv.reader(ft)
     result=list(reader)
     for i in result[1:]:
          a=i[0].split()
          for j in a:
               if dictionary.get(j)==None:
                    dictionary[f'{j}']=1
               else:dictionary[f'{j}'] = int(dictionary[f'{j}'])+1
# dictionary
# dictMerged2 = dict( dict1, **dict2 )
```
